```python
def opinion_dic(dic):
    """
    aspect = key + fundamental + noun_inter
    sentimental = positive + negative
    """
    opinion_dic = {}
    for key,value in dic.items():
        aspect = delete_duple_(value["key"] + value["fundamental"] + value["noun_inter"])
        sentimental = delete_duple_(value["positive"]+value["negative"])
        opinion_dic[key] = {"aspect":aspect,"sentiment":sentimental,"positive":value["positive"],"negative":value["negative"]}
    return opinion_dic
```

```python
def delete_duple_(list_):
    tass = []
    for i in list_:
        if i in tass:
            pass
        else:
            tass.append(i)
    return tass
```

```python
def score_dic(dic,target_col):
    inter_ = []
    for key,value in opi_dic.items():
        exten_ = []
        [exten_.extend(i) for i in value[target_col]]
        inter_.extend(list(set(exten_)))
    
    "socore =  Counter의 역수 값."
    score_dic = {}
    score = [(i[0],1/i[1]) for i in Counter(inter_).most_common()]
    for s in score:
        score_dic[s[0]] = s[1]
    return score_dic
```

```python
def combine_dicts(opi_dic,sco_dic,target_col):
    opi_dic2 = {}
    if target_col == "sentiment":
        for key,value in opi_dic.items():
            exten_ = []
            [exten_.extend(i) for i in value["sentiment"]]
            exten_1 = list(set(exten_))
            exten_2 = [(ex,sco_dic[ex]) for ex in exten_1]
            opi_dic2[key] = {"aspect":value["aspect"],"sentiment":exten_2,
                             "positive":value["positive"],"negative":value["negative"]}
    elif target_col == "aspect":
        for key,value in opi_dic.items():
            opi_dic2[key] = {"aspect":value["aspect"],"sentiment":value["sentiment"],
                             "positive":value["positive"],"negative":value["negative"]}
    return opi_dic2
```

```python
def attr_score(Sens,Attr):
    index = []
    attrs = []
    for attr_ in Attr:
        if attr_[0] in Sens:
            index.append(Sens.index(attr_[0]))
            attrs.append(attr_)
        else:
            pass

    return index,attrs
```

```python
def out_dic(combi_dic,target_col,sen_test,threshold):
    senti_out = {}
    score_out = {}
    score_total_out = {}
    for key,value in combi_dic.items():
        score_ = attr_score(sen_test,value[target_col])[1]
        if target_col =="sentiment":
            score_thres = [sco_ for sco_ in score_ if sco_[1] >= threshold]
        elif target_col == "aspect":
            score_thres = score_

        for sth in score_thres:
            if sth[0] in list(senti_out.keys()):
                senti_out[sth[0]] = senti_out[sth[0]] + [key]
            else:
                senti_out[sth[0]] = [key]
        
        try:
            if sum([sco[1] for sco in score_]) != 0:
                score_total_out[key] = sum([sco[1] for sco in score_])
            if sum([sco[1] for sco in score_thres]) != 0:
                score_out[key] = sum([sco[1] for sco in score_thres])
        except:
            pass
    senti_out_2 = {}
    for key,value in senti_out.items():
        value = list(set(value))
        senti_out_2[key] = value
    return senti_out_2,score_out,score_total_out
```

```python
def attr_split(Sens,index):
    out_ = []
    buffer_ = []
    for i in index:
        candi_ = Sens[0:i+1]
        inter = list(set(candi_).intersection(set(buffer_)))
        [candi_.remove(remo) for remo in inter]
        out_.append(candi_)
        buffer_ = Sens[0:i+1]
    return out_
```

```python
def corre_df(measures,finder,version):
    corr_dic = {}
    if version == "pmi":
        tq = tqdm(finder.score_ngrams(measures.pmi),desc="Making {}".format(version))
    elif version == "raw_freq":
        tq = tqdm(finder.score_ngrams(measures.raw_freq),desc="Making {}".format(version))
    elif version == "likelihood_ratio":
        tq = tqdm(finder.score_ngrams(measures.likelihood_ratio),desc="Making {}".format(version))
    else:
        tq = tqdm(finder.score_ngrams(measures.likelihood_ratio),desc="Making {}".format(version))      
    t = 0
    for i in tq:
        #print(i)
        target_ = i[0][0]
        link_ = i[0][1]
        score_ = i[1]
        corr_dic[t] = [target_,link_,score_]
        t += 1
    corr_df = pd.DataFrame(corr_dic).T.rename(columns={0:"target_word",1:"link_word",2:"score"})
    return corr_df
```

```python
def correlation(df_):
    tagged_words = []
    for sen in df_["sentence"]:
        tagged_words.append(sen)
    measures = collocations.BigramAssocMeasures()
    finder = collocations.BigramCollocationFinder.from_documents(tagged_words)
    pmi_df = corre_df(measures,finder,"pmi")
    freq_df = corre_df(measures,finder,"raw_freq")
    like_df = corre_df(measures,finder,"likelihood_ratio")
    return pmi_df,freq_df,like_df
```

```python
pmi_df,freq_df,like_df = correlation(df_)
```
